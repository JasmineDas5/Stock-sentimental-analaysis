```python
df=pd.DataFrame(columns=['Date','Top1','Top2','Top3','Top4','Top5','Top6','Top7','Top8','Top9','Top10','Top11','Top12','Top13','Top14','Top15'])
def get_data():
  url='https://economictimes.indiatimes.com/markets/stocks/news'
  response=requests.get(url)
  soup=BeautifulSoup(response.content,'html.parser')
  date=soup.find('span',attrs={"class":"dib time"})
  for d in soup.find_all('div',attrs={"class":"tabdata"}):
    top1=d.find('a',attrs={"data-orefid":"p2146843_1"})
    top2=d.find('a',attrs={"data-orefid":"p2146843_2"})
    top3=d.find('a',attrs={"data-orefid":"p2146843_3"})
    top4=d.find('a',attrs={"data-orefid":"p2146843_4"})
    top5=d.find('a',attrs={"data-orefid":"p2146843_5"})
    top6=d.find('a',attrs={"data-orefid":"p2146843_6"})
    top7=d.find('a',attrs={"data-orefid":"p2146843_7"})
    top8=d.find('a',attrs={"data-orefid":"p2146843_8"})
    top9=d.find('a',attrs={"data-orefid":"p2146843_9"})
    top10=d.find('a',attrs={"data-orefid":"p2146843_10"})
    top11=d.find('a',attrs={"data-orefid":"p2146843_11"})
    top12=d.find('a',attrs={"data-orefid":"p2146843_12"})
    top13=d.find('a',attrs={"data-orefid":"p2146843_13"})
    top14=d.find('a',attrs={"data-orefid":"p2146843_14"})
    top15=d.find('a',attrs={"data-orefid":"p2146843_15"})
    all=[]
    if date is not None:
      all.append(date.text[0:17])
    else:
      all.append(" ")
    if top1 is not None:
      all.append(top1.text)
    else:
      all.append(" ")
    if top2 is not None:
      all.append(top2.text)
    else:
      all.append(" ")
    if top3 is not None:
      all.append(top3.text)
    else:
      all.append(" ")
    if top4 is not None:
      all.append(top4.text)
    else:
      all.append(" ")
    if top5 is not None:
      all.append(top5.text)
    else:
      all.append(" ")
    if top6 is not None:
      all.append(top6.text)
    else:
      all.append(" ")
    if top7 is not None:
      all.append(top7.text)
    else:
      all.append(" ")
    if top8 is not None:
      all.append(top8.text)
    else:
      all.append(" ")
    if top9 is not None:
      all.append(top9.text)
    else:
      all.append(" ")
    if top10 is not None:
      all.append(top10.text)
    else:
      all.append(" ")
    if top11 is not None:
      all.append(top11.text)
    else:
      all.append(" ")
    if top12 is not None:
      all.append(top12.text)
    else:
      all.append(" ")
    if top13 is not None:
      all.append(top13.text)
    else:
      all.append(" ")
    if top14 is not None:
      all.append(top14.text)
    else:
      all.append(" ")
    if top15 is not None:
      all.append(top15.text)
    else:
      all.append(" ")
    df.loc[len(df)]=all
  return df
```

```python
new_df['Date'] = pd.to_datetime(new_df['Date'])
train = new_df[new_df['Date'] < '2023-12-02']
test = new_df[new_df['Date'] >= '2023-12-02']
```

```python
data= train.iloc[:,1:16]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# RENAMING COLUMN NAME FOR EASE OF ACCESS
list1=[i for i in range(15)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
```

```python
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
for index in new_Index:
    data[index] = data[index].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
' '.join(str(x) for x in data.iloc[1,0:16])
```

```python
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:16]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector= CountVectorizer(ngram_range=(2,2))
traindataset= countvector.fit_transform(headlines) # CONVERTING ALL THE HEADLINES INTO VECTORS
```

```python
traindataset[0]
```

```python
random_classifier= RandomForestClassifier(n_estimators=10000,criterion='entropy')
random_classifier.fit(traindataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,1:16]))
test_dataset= countvector.transform(test_transform)
predictions= random_classifier.predict(test_dataset)
```

```python
predictions
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
matrix= confusion_matrix(test["Label"],predictions)
print(matrix)
score= accuracy_score(test["Label"],predictions)
print(score)
report= classification_report(test['Label'],predictions)
print(report)
```

```python
# USING RANDOM FOREST CLASSIFIER WITH TF-IDF VECTORIZER
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
tfidf= TfidfVectorizer(ngram_range=(2,2))
traindataset= tfidf.fit_transform(headlines)
```

```python
random_classifier= RandomForestClassifier(n_estimators=200,criterion='entropy')
random_classifier.fit(traindataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,1:16]))
test_dataset= tfidf.transform(test_transform)
predictions= random_classifier.predict(test_dataset)
matrix= confusion_matrix(test["Label"],predictions)
print(matrix)
score= accuracy_score(test["Label"],predictions)
print(score)
report= classification_report(test['Label'],predictions)
print(report)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming headlines is the preprocessed feature derived from train
X_train = headlines
y_train = train['Label']

# Assuming test_transform is the preprocessed feature derived from the test dataset
X_test = test_transform
y_test = test['Label']  # Assuming 'Label' is the target variable in the test dataset

# Vectorize the headlines using CountVectorizer
vectorizer = CountVectorizer(max_features=5000)
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Build the ID3 Decision Tree model
model = DecisionTreeClassifier()
model.fit(X_train_vectorized, y_train)

# Make predictions on the test set
predictions = model.predict(X_test_vectorized)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy}")

# Get the classification report
report = classification_report(y_test, predictions)
print("Classification Report:\n", report)
```
